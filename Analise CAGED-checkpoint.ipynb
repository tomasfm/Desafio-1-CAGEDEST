```python
man = caged[caged.Sexo == 1]

wom = caged[caged.Sexo == 2]
```

```python
man.Salario_Mensal.plot.hist(bins=100)

wom.Salario_Mensal.plot.hist(bins=100)
```

```python
man['Log_Salario_Mensal'] = np.log(man['Log_Salario_Mensal'])
man['Log_Salario_Mensal'].plot.hist(bins=200)

wom['Log_Salario_Mensal'] = np.log(wom['Log_Salario_Mensal'])
wom['Log_Salario_Mensal'].plot.hist(bins=200)
```

```python
# Plotando uma distribuição do log normalizado
man['Normal_Log_Salario_Mensal'] = ( man['Log_Salario_Mensal'] - 
                                    man['Log_Salario_Mensal'].mean() ) / man['Log_Salario_Mensal'].std()
man['Normal_Log_Salario_Mensal'].plot.hist(200)

wom['Normal_Log_Salario_Mensal'] = ( wom['Log_Salario_Mensal'] - 
                                    wom['Log_Salario_Mensal'].mean() ) / wom['Log_Salario_Mensal'].std()
wom['Normal_Log_Salario_Mensal'].plot.hist(200)
```

```python
# definindo os dados
x_man = man['Log_Salario_Mensal']
# calculando a média 
mu_man = np.mean(x_man)
# calculando o desvio=padrão (utilizando o padrão para o Python)
std_man = np.std(x_man)


# definindo os dados
x_wom = wom['Log_Salario_Mensal']
# calculando a média 
mu_wom = np.mean(x_wom)
# calculando o desvio=padrão (utilizando o padrão para o Python)
std_wom = np.std(x_wom)
mu_man,std_man,mu_wom,std_wom
```

```python
# p-value para distribuição normal
scipy.stats.norm(mu_man, std_man).pdf(2*std_man)
```

```python
# z-score para distribuição normal
scipy.stats.norm.interval(alpha=0.95,loc=mu_man,scale=std_man)
```

```python
tes = mu_wom+2*std_wom
tes
```

```python
# Utilizando a estatística t-student
Z_Score_Inf = scipy.stats.t.interval(alpha=0.95,
                       df=(len(man['Log_Salario_Mensal'])-1),
                       loc=mu_man,scale=std_man
                      )[0] # df=Degrees of Freedom
Z_Score_Inf
```

```python
Z_Score_Sup = scipy.stats.t.interval(alpha=0.95,
                       df=(len(man['Log_Salario_Mensal'])-1),
                       loc=mu_man,scale=std_man
                      )[1] # df=Degrees of Freedom
Z_Score_Sup
```

```python
man[man['Log_Salario_Mensal']>Z_Score_Sup]
```
